```python
# Plot the Image and its histogram + cdf of the original image I chose a color picture of your own choice
# Note that it is a color image, so it will have three different histograms

### BEGIN SOLUTION

### END SOLUTION

import numpy as np
from skimage import io, color, exposure
import matplotlib.pyplot as plt





```

```python
I = io.imread("k1.jfif")
print(I.shape)
```

```python
I
```

```python
io.imshow(I)
plt.axis("off")
```

```python
hist, bins_hist = exposure.histogram(I.ravel(), nbins=256, normalize=True) # pdf

cdf, bins_cdf = exposure.cumulative_distribution(I.ravel(), nbins=256) # cdf
```

```python
# Split into RGB channels
R, G, B = I[:, :, 0], I[:, :, 1], I[:, :, 2]

# Calculate histograms and CDFs
hist_R, bins_R = np.histogram(R.flatten(), bins=256, range=[0, 256])
hist_G, bins_G = np.histogram(G.flatten(), bins=256, range=[0, 256])
hist_B, bins_B = np.histogram(B.flatten(), bins=256, range=[0, 256])

cdf_R = np.cumsum(hist_R) / hist_R.sum()
cdf_G = np.cumsum(hist_G) / hist_G.sum()
cdf_B = np.cumsum(hist_B) / hist_B.sum()

# Plotting
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(I)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Histograms")
plt.plot(hist_R, color='red', label='Red')
plt.plot(hist_G, color='green', label='Green')
plt.plot(hist_B, color='blue', label='Blue')
plt.xlim([0, 256])
plt.legend()

plt.subplot(2, 2, 3)
plt.title("CDFs")
plt.plot(cdf_R, color='red', label='Red')
plt.plot(cdf_G, color='green', label='Green')
plt.plot(cdf_B, color='blue', label='Blue')
plt.xlim([0, 256])
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# make the image grascale . Plot the Image and its histogram + cdf of the grayscale image I_gray

### BEGIN SOLUTION

### END SOLUTION
# Convert to grayscale
I_gray = color.rgb2gray(I)
hist, bins = np.histogram(I_gray.flatten(), bins=256, range=[0, 1])
cdf = np.cumsum(hist) / hist.sum()

```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Grayscale Image")
plt.imshow(I_gray, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Histogram")
plt.plot(hist, color='black')
plt.xlim([0, 250])

plt.subplot(2, 2, 3)
plt.title("CDF")
plt.plot(cdf, color='black')
plt.xlim([0, 250])

plt.tight_layout()
plt.show()
```
